```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
plt.title('젠슨황')
plt.xlabel('x축', color='red')
plt.ylabel('y축', color='blue', loc='top')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [50, 60 ,70 ,80]
```

```python
plt.figure(dpi=50)
plt.plot(x, y, linewidth=1, marker='o', mec='red', mfc='yellow', ms=10, linestyle='-.', color='g') 
plt.xlabel('공부시간')
plt.ylabel('점수')
for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx]+0.7, txt)
plt.show()
```

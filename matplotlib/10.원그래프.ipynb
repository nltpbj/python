```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #맑은 고딕
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시 마이너스 글자가 깨지는 현상을 해결
import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
plt.pie(values, labels=labels, autopct=' %.1f%%',
        startangle=90, counterclock=False)
plt.show()
```

```python
explode=[0.02] * len(values)
explode
```

```python

plt.pie(values, labels=labels, autopct=' %.1f%%',
        startangle=90, counterclock=False, explode=explode)
plt.legend(loc=(1.2, 0.3), title='선택언어')
plt.show()
```

```python
colors=['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF'] 
```

```python
plt.pie(values, labels=labels, autopct=' %.1f%%',
        startangle=90, counterclock=False, explode=explode, colors=colors)
plt.legend(loc=(1.2, 0.3), title='선택언어')
plt.show()
```

```python
def custom_autopct(pct):
    if pct >=10:
        return '{:.1f}%'.format(pct)
    else: 
        return ''
```

```python
plt.pie(values, labels=labels, autopct=' %.1f%%',
        startangle=90, counterclock=False, explode=explode, colors=colors,
        wedgeprops={'width':0.8}, pctdistance=0.7)
plt.legend(loc=(1.2, 0.3), title='선택언어')
plt.show()
```

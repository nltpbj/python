```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #맑은 고딕
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시 마이너스 글자가 깨지는 현상을 해결

import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
x = list(df['영어'])
y = list(df['수학'])
sizes = list(df['학년'] * 500)
sizes
```

```python

plt.figure(figsize=(5, 5))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
for idx, no in enumerate(df['이름']):
  plt.text(x[idx], y[idx]+1, no, size=8, ha='center', color='green')
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)
plt.show()
```

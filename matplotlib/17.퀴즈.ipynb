```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #맑은 고딕
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시 마이너스 글자가 깨지는 현상을 해결
```

```python
data = {
 '영화' : ['명량', '극한 직업', '신과 함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번 방의 선물', '암살'],
 '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
 '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], #(단위 : 만 명)
 '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['영화'], df['평점'])
plt.xticks(rotation=90)
plt.title('국내영화 평점')
plt.ylabel('평점')
plt.show()
```

```python
df_groub = df.groupby('개봉 연도')[['관객 수', '평점']].mean()
df_groub
```

```python
plt.plot(df_groub.index, df_groub['평점'], marker='x')
plt.xticks([2005, 2010, 2015, 2020])
plt.show()
```

```python

data = {
    '날짜': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01', '2021-06-01', '2021-07-01', '2021-08-01'],
    '비트코인 가격 (USD)': [29000, 33000, 45000, 59000, 49000, 35000, 40000, 47000]
}

df = pd.DataFrame(data)
df['날짜'] = pd.to_datetime(df['날짜'])

plt.figure(figsize=(10, 6))
plt.plot(df['날짜'], df['비트코인 가격 (USD)'], marker='o', color='tab:blue')

plt.title('화성가보자', fontsize=16)
plt.xlabel('날짜', fontsize=14)
plt.ylabel('비트코인 가격 (USD)', fontsize=14)

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/LinearRegressionData.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X, y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

```python
y_pred = reg.predict(X)
y_pred
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='green')
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
dataset = pd.read_csv('data/LinearRegressionData.csv')
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.predict([[3.2]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train, len(X_train)
```

```python
X_test, len(X_test)
```

```python
reg.fit(X_train, y_train)
```

```python
y_test, len(y_test)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.title('Score by hours (test data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
sr = SGDRegressor()
sr.fit(X_train, y_train)
```

```python
sr.score(X_train, y_train), reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import SGDRegressor
sr = SGDRegressor(max_iter=1000, eta0=0.001, random_state=0, verbose=1)
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, sr.predict(X_train), color='green')
plt.title('Score by hours (test data: SGD)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
